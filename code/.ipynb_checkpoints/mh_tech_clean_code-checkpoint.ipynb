# Data Cleaning

## 2018  

```python
df2018 = pd.read_csv('../data/survey_2018.csv')
```

```python
#make columns headers lowercase
df2018.columns = map(str.lower, df2018.columns)
#remove html
df2018.rename(columns=lambda colname: re.sub('</\w+>', '', colname), inplace=True)
df2018.rename(columns=lambda colname: re.sub('<\w+>', '', colname), inplace=True)
```

```python
len(df2018.columns)
```

```python
df2018.count()
```

```python
#drop self-employed with no previous employers
df2018 = df2018.drop(df2018[(df2018['are you self-employed?'] == 1)].index)
df2018.describe()
```

```python
#rename columns
df2018_rename = df2018.rename(columns={
     'how many employees does your company or organization have?': 'no_employees',
     'is your employer primarily a tech company/organization?':'tech_company',
     'does your employer provide mental health benefits as part of healthcare coverage?': 'benefits',
     'do you know the options for mental health care available under your employer-provided health coverage?':'care_options',
     'has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?': 'wellness_program',
     'does your employer offer resources to learn more about mental health disorders and options for seeking help?': 'seek_help', 
     'is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources provided by your employer?': 'anonymity',
     'if a mental health issue prompted you to request a medical leave from work, how easy or difficult would it be to ask for that leave?': 'leave',
     'would you feel comfortable discussing a mental health issue with your direct supervisor(s)?': 'supervisor',
     'would you feel comfortable discussing a mental health issue with your coworkers?': 'coworkers',
     'have you ever sought treatment for a mental health disorder from a mental health professional?' : 'treatment',
     'do you have a family history of mental illness?' : 'family_history',
     'would you be willing to bring up a physical health issue with a potential employer in an interview?': 'phys_health_interview',
     'would you bring up your mental health with a potential employer in an interview?' : 'mental_health_interview',
     'have you observed or experienced an unsupportive or badly handled response to a mental health issue in your current or previous workplace?': 'obs_consequence',
     'what is your age?' : 'age',
     'what is your gender?' : 'gender',                                     
    })
```

```python
#select columns
df2018_clean = df2018_rename[[
     'no_employees',
     'tech_company',
     'benefits',
     'care_options',
     'wellness_program',
     'seek_help', 
     'anonymity',
     'leave',
     'supervisor',
     'coworkers',
     'treatment',
     'family_history',
     'phys_health_interview',
     'mental_health_interview',
     'obs_consequence',
     'age',
     'gender', 
]]
df2018_clean.head()
```

```python
#normalize response to observer consequence to yes if experience or observed
df2018_clean['obs_consequence'] = df2018_clean['obs_consequence'].replace({'Yes, I experienced': 'Yes', 'Yes, I observed': 'Yes'})
df2018_clean['leave'] = df2018_clean['leave'].replace({'Somewhat difficult': 'Difficult', 
                                                      'Very easy': 'Easy', 'Somewhat easy': 'Easy'})
```

```python
df2018_clean.shape
```

## 2014

```python
df2014 = pd.read_csv('../data/survey_2014.csv')
#make columns lowercase
df2014.columns = map(str.lower, df2014.columns)
list(df2014)
```

```python
len(df2014.columns)
```

```python
df2014_clean = df2014[[ 
     'no_employees',
     'tech_company',
     'benefits',
     'care_options',
     'wellness_program',
     'seek_help', 
     'anonymity',
     'leave',
     'supervisor',
     'coworkers',
     'treatment',
     'family_history',
     'phys_health_interview',
     'mental_health_interview',
     'obs_consequence',
     'age',
     'gender',  
]]
df2014_clean.head()
```

```python
df2014_clean['tech_company'] = df2014_clean['tech_company'].replace({'Yes': 1, 'No': 0})
df2014_clean['care_options'] = df2014_clean['care_options'].replace({'Not sure':'I am not sure'})
df2014_clean['treatment'] = df2014_clean['treatment'].replace({'Yes': 1, 'No': 0})
df2014_clean['leave'] = df2014_clean['leave'].replace({'Somewhat difficult': 'Difficult', 'Very difficult': 'Difficult',
                                                      'Very easy': 'Easy', 'Somewhat easy': 'Easy'})
df2014_clean = df2014_clean.replace({"Don't know": "I don't know"})
df2014_clean.head()
```

```python
clean_aggr = pd.concat([df2014_clean, df2016_clean, df2017_clean, df2018_clean], sort=False)
clean_aggr.count()
```

```python
aggr_survey = clean_aggr.dropna(how = 'any') 
aggr_survey.count()
```

```python
aggr_survey['gender'] = aggr_survey['gender'].apply(str.lower)
```

```python
# categorize identify with male
aggr_survey = aggr_survey.replace([
    'male', 'male ','ostensibly male', 'male, born with xy chromosoms', 'malel', 'm', 'male (or female, or both)', 
    'cis-male',  'cis male', 'cis male ', 'man', 'demiguy', 'cisgender male', 'swm', 'mail', 
    'cis hetero male', 'male (cis)', 'dude', 'male-ish','male.', 
    'male 9:1 female, roughly', 'sex is male', 'cisdude', 'cis man', 'maile', 
    'something kinda male?', 'mal', 'make', 'guy (-ish) ^_^', 'msle', 'malr', 
    'ostensibly male, unsure what that really means'], 'M')
#categorize identify with female
aggr_survey = aggr_survey.replace([
    'female', 'female ','woman', 'f', 'i identify as female', 'i identify as female.','*shrug emoji* (f)', 'my sex is female.',
    'female/gender non-binary.', 'she/her/they/them', 'gender non-conforming woman', 'female assigned at birth ',
    'female (props for making this a freeform field, though)', 'female/woman', ' female', 'female-bodied; no feelings about gender',
    'cisgender female', 'cis female', 'cis female ', 'female (cisgender)', 'female (cis)', 'female (cis) ',
    'cis-female', 'cisgendered woman', 'cis woman', 'cis-woman', 'cis-female/femme',
    'femalw', 'femail', 'f, cisgender', 'fm', 'female or multi-gender femme', 'fem', 'femake', ],'F')

#categorize identify with non-binary
aggr_survey = aggr_survey.replace([
    'agender', 'androgynous', 'androgyne', 'nonbinary', 'agender/genderfluid','non binary', 'non-binary', 'nb', 
    'nonbinary/femme', 'genderqueer/non-binary', 'trans man', 'transgender',  'trans woman', 'trans female', 
    'female (trans)','trans-female', 'transfeminine', 'other/transfeminine','transgender woman','genderfluid', 
    'genderqueer' , 'genderqueer demigirl','genderfluid (born female)', 'genderqueer woman','bigender',
    'genderflux demi-girl', 'uhhhhhhhhh fem genderqueer?', 'transitioned, m2f', 'male leaning androgynous',
    'nb masculine', 'male/genderqueer', 'human', 'enby','male (trans, ftm)', 'fluid', 'queer', 'queer/she/they',
    'a little about you', 'none of your business', 'sometimes', 'neuter', 'p','nah', 'all','n/a', 'none', 
    'other', 'god king of the valajar', 'contextual', '\-', ], 'nonbinary')


```

## Visualization

```python
#prep catagorical data
aggr_cat = aggr_survey.drop(columns =["tech_company", "treatment", "age"])
aggr_cat.head()
```

```python
features = list(aggr_cat)
for ll in features:
    aggr_cat[ll] = pd.Categorical(aggr_cat[ll]).codes
aggr_cat.head(5)
```

```python
#select numerical data
aggr_num = aggr_survey[[ 
     "tech_company", "treatment", "age"]]
aggr_num.head()
```

```python
#concatenate the two dataframes for visualization
aggr_visual = pd.concat([aggr_cat, aggr_num], axis=1)
aggr_visual.head()
```

```python
#plot correlation matrix visual
import matplotlib.pyplot as plt

plt.matshow(aggr_visual.corr())
plt.show()
```
